```python
df['is_expense'] = np.where(df.amount < 0, 1, 0)
```

```python
df['date'] = pd.to_datetime(df.date, infer_datetime_format=True)
```

```python
fig, ax = plt.subplots(figsize=(10,13))

sns.countplot(y='category', hue='homeownership', data=df, order=df.category.value_counts().iloc[:20].index, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(10,13))

sns.countplot(y='category', hue='homeownership', data=df[df.is_expense == 1], order=df.category.value_counts().iloc[:20].index, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(10,13))

sns.countplot(y='category', hue='homeownership', data=df[df.is_expense == 0], order=df.category.value_counts().iloc[:20].index, ax=ax)
```
